```python
def imshow(imgnumpy: np.ndarray):
    plt.imshow(imgnumpy)
```

```python
class ToNumpy(torch.nn.Module):
    def __init__(self, inplace=False):
        super().__init__()
        self.inplace = inplace

    def forward(self, img):
        return np.array(img)
```

```python
tf_tonumpy = ToNumpy()
image_numpy_torch = tf_tonumpy(image)
```
